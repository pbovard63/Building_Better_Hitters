For this second test, I'd like to remove a few numerical features that have lower correlation: Zone swing rate and Sprint Speed.  Additionally, I am not going to include positions (categorical data) here.  

My hypothesis is that positions were being falsely attributed to offensive success or lack thereof - i.e., model is using position to say a player is performing poorly offensively because they play a certain position, while that is likely not true (using my domain knowledge).  

I'll do this in two steps: 

1. a model with just positions removed
2. a model with positions and lesser-correlated stats removed

Removing unnecessary columns in that case:

```python
model_df = stats_df.drop(columns=['Player', 'Year', 'PA', 'G', 'Player Link_x', 'Player_Year_ID', 'Position'])
```

```python
model_df.head()
```

## New Model 1: Just Positions Removed (No Categorical Data)

### Simple Linear Regression, with Cross Validation:

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
```

```python
X, y = model_df.drop('wRC+',axis=1), model_df['wRC+']
```

Splitting Data 80/20:

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=.2, random_state=10)
```

```python
X_train_val.shape, X_test.shape
```

```python
y_train_val.shape, y_test.shape
```

Since there is no categorical data, no One Hot Encoding needed. On to K-Folds:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge #ordinary linear regression + w/ ridge regularization
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
X_train_val, y_train_val = np.array(X_train_val), np.array(y_train_val)
```

```python
#run the CV, doing 10 folds
kf = KFold(n_splits=10, shuffle=True, random_state = 2)
cv_lm_r2s, cv_lm_reg_r2s = [], [] #collect the validation results for both models

for train_ind, val_ind in kf.split(X_train_val,y_train_val):
    
    X_train, y_train = X_train_val[train_ind], y_train_val[train_ind]
    X_val, y_val = X_train_val[val_ind], y_train_val[val_ind] 
    
    #simple linear regression
    lm = LinearRegression()

    lm.fit(X_train, y_train)
    cv_lm_r2s.append(lm.score(X_val, y_val))

print('Simple regression scores: ', cv_lm_r2s)
print(f'Simple mean cv r^2: {np.mean(cv_lm_r2s):.3f} +- {np.std(cv_lm_r2s):.3f}')
```

Scoring on the test data frames:

```python
lm.fit(X_train_val, y_train_val)
```

```python
lm.score(X_test, y_test)
```

Hmm, well that's worse... (was 0.3939786097719923 on first_model)

```python
sorted(list(zip(X_test.columns, np.round(lm.coef_, 3))), key=lambda x: x[1])
```
